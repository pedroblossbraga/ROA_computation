# Region Of Attraction (ROA) computation


The dual of the problem can be formulated with the following optimization problem
\begin{align}
\inf & \int_X w(x)d \lambda(x) & \\
\text{s.t. } & \mathcal{L} v(t,x,u) \leq 0, & \forall(t,x,u) \in [0,T] \times X \times U \\
& w(x) \geq v(0,x) + 1, & \forall x \in X \\
& v(t,x) \geq 0, & \forall (t,x) \in [0,T] \times X_T \\
& w(x) \geq 0, & \forall x \in X \\
\end{align}

## Example 1: Univariave Cubic Dynamics
\begin{align}
\dot{x} = x(x-0.5)(x+0.5) \\
X = [-1,1], T = 100, X_T = [-0.01, 0.01]
\end{align}
- The ROA is analytically found as $X_0 = [-0.5, 0.5]$. 
- Show polynomial approximations for degrees $d \in \{4, 8, 16, 32 \}$

## ROA computation

```julia
using Printf
using DynamicPolynomials
using LinearAlgebra
using JuMP
using SumOfSquares
using SCS
using MosekTools
using LazySets
using Plots
using IntervalArithmetic

function f(x)
    return x * (x - 0.5) * (x + 0.5)*10
end

function compute_roa(d, T, α, β, a, b, method="Mosek")
    """
    ----------------------------------------------------------------------
    Parameters:
    - d: degree of polynomial
    - T: terminal time
    - [α,β]: interval close to the stable equilibrium point
    - [a,b]: interval defining the set X
    Returns:
    - v and w solutions
    ----------------------------------------------------------------------
    """
    # Define the model
    if method == "SCS"
        model = Model(SCS.Optimizer) # Splitting Conic Solver, slow!
    else
        model = Model(Mosek.Optimizer)
    end
    
    @polyvar t x
    
    #----------------------------------------------------------------------
    # semialgebraic sets
    
    # [0,T] × X
    K = @set -t*(t-T) ≥0 && -(x-a)*(x-b) ≥0
    X = @set -(x-a)*(x-b) ≥0
    K_terminal = @set -t*(t-T) ≥0 && -(x-α)*(x-β) ≥0
    
    #----------------------------------------------------------------------
    # Define the Lyapunov function v(x) and the polynomial w(x)
    
    @variable(model, v, Poly(monomials([t, x], 0:d)))
    @variable(model, w, Poly(monomials(x, 0:d)))
    
    # Evaluate v(0,x)
    v_0 = subs(v, t=>0)
    
    #----------------------------------------------------------------------
    # Constraints
    
    # Defining L operator Lv = grad v * f + dv/dt
    Lv = differentiate(v, x) * f(x) + differentiate(v, t)
    
    # Lv(t,x) ≤ 0, ∀(t,x) ∈ [0,T] × X
    @constraint(model, -Lv in SOSCone(), domain=K)
    
    # w(x) ≥ v(0,x) + 1, ∀x ∈ X
    @constraint(model, w - v_0 - 1 in SOSCone(), domain=X)
    
    # v(T,x) ≥ 0, ∀(t,x) ∈ [0,T] × X_T
    @constraint(model, v in SOSCone(), domain=K_terminal)
    
    # w(x) ≥ 0, ∀x ∈ X
    @constraint(model, w in SOSCone(), domain=X)
    
    #----------------------------------------------------------------------
    # Objective: Minimize the integral of w over X (using fundamental Thm of calculus)
    
    @objective(model, Min, sum(w.a[i] * (b^(i) - a^(i)) / i for i in 1:d+1))
    
    #----------------------------------------------------------------------
    # Solve the problem
    optimize!(model)
    
    # Extract and return the solutions
    v_sol = value(v)
    w_sol = value(w)

    return v_sol, w_sol, x, t
end


function plot_results(x_vals, w_vals, d, method, w0)
    # Plot the polynomial approximation w(x)
    plot(x_vals, w_vals, label="w(x)", lw=2)
    xlabel!("x")
    ylabel!("Value")
    title!("Polynomial Approximation for ROA d=$(d) ($(method))")
    
    # Plot the ROA region
    x_roa = [-0.5, -0.5, 0.5, 0.5]
    y_roa = [0, 1, 1, 0]
    plot!(
        x_roa, y_roa,
        linestyle=:dash,
        seriescolor=:gray,
        label="ROA [-0.5, 0.5]"
    )
    scatter!([0], [w0], color=:red, label="w(0)=$(@sprintf("%.5f", w0))")
end

# ----------------------------------------------
function run_simulation(;
        d = 8,
        T = 100,
        α = -0.01,
        β = 0.01,
        a = -1,
        b = 1,
        method = "Mosek",
        verbose=true
        )
    v_sol, w_sol, x, t = compute_roa(d, T, α, β, a, b, method)
    
    if verbose
        # Print the results
        println("Lyapunov function coefficients v(x): \n$v_sol\n")
        println("Polynomial coefficients w(x): \n$w_sol\n")
    end

    x_vals = range(a, b, length=100)

    # Evaluate w(x) for the given range
    w_vals = [w_sol(x=>point) for point in x_vals]
    
    w0 = w_sol(0)
    println("w(0)=", w_sol(0))

    # Plot the results
    plot_results(x_vals, w_vals, d, method, w0)
end

# run_simulation()
```

```julia
function gather_results(;d=8, T=100, α=-0.01, β=0.01, a=-1, b=1, method="Mosek", verbose=true)
    v_sol, w_sol, x, t = compute_roa(d, T, α, β, a, b, method)
    
    if verbose
        # Print the results
        println("Lyapunov function coefficients v(x): \n$v_sol\n")
        println("Polynomial coefficients w(x): \n$w_sol\n")
    end

    x_vals = range(a, b, length=100)

    # Evaluate w(x) for the given range
    w_vals = [w_sol(x=>point) for point in x_vals]

    w0 = w_sol(0)
    return x_vals, w_vals, d, method, w0
end

function make_roa_box(v)
    x_roa = [-v, -v, v, v]
    y_roa = [0, 1, 1, 0]
    return x_roa, y_roa
end

function plot_all_results(results, output_filename)
    n = length(results)
    plot_grid = plot(layout=(1, n), size=(1200, 400))
    
    roa_box_colors=[:green, :orange, :red]
    
    for (i, (x_vals, w_vals, d, method, w0)) in enumerate(results)
        use_legend=:outerbottom
        
        plot!(plot_grid[i], x_vals, w_vals, label="w(x)", lw=2, 
            seriescolor=:blue,
            legend=use_legend,
            xlabel="x",
            title="d=$d, w(0)=$(@sprintf("%.5f", w0))"
        )
        xlabel!(plot_grid[i], "x")
        ylabel!(plot_grid[i], "Value")

        for (c, v) in enumerate([0.5, 0.25, 0.1])
            x_roa, y_roa = make_roa_box(v)
            plot!(
                plot_grid[i], x_roa, y_roa,
                linestyle=:dash, linewidth=2,
                seriescolor=roa_box_colors[c],
                label="[$(-v), $(v)]",
                use_legend=use_legend,
            )
        end
    end

    savefig(plot_grid, output_filename)
    display(plot_grid)
end
```

```julia
using DifferentialEquations
using Plots

# Define the differential equation
function f(dx, x, params, t)
    dx .= -10 .* x .* (x .- 0.5) .* (x .+ 0.5)
end

# Initial condition
x0 = [0.25]

# Time span
tspan = (0.0, 10.0)


plot_grid = plot(
    # layout=(1, n),
    size=(700, 400)
)

# Define the problem
prob = ODEProblem(f, x0, tspan)

# Solve the problem
sol = solve(prob)

# Plot the solution
plot!(sol, xlabel="Time", ylabel="x(t)", legend=false, 
    title="Solution to the ODE", 
    # size=(800, 300)
)

# Save the plot
savefig("ode_solution_19jul.png")
display(plot_grid)
```

## Univariate Cubic Dynamics plot

```julia
using LaTeXStrings
using Plots

# Define the differential equation
function f(x)
    return x * (x - 0.5) * (x + 0.5)
end

# Define the differential equation
function f(dx, x, params, t)
    return x * (x - 0.5) * (x + 0.5)
end

# Solve for equilibrium points using DifferentialEquations.jl
equilibria = find_zeros(f, (-1.0, 1.0))

println("equilibrium points: $equilibria")

# Define a range of x values
x = range(-1, 1, length=400)

# Compute the corresponding y values (dx/dt = f(x))
y = f.(x)

# Create the plot object
p = plot(x, y, xlabel="x", ylabel=L"\dot{x}", 
    legend=true, label=L"\dot{x}",
    title=L"Direction Field for $\dot{x} = x(x-0.5)(x+0.5)$")

unstable_eq = [-0.5, 0.5]
stable_eq = [0]
scatter!(unstable_eq, [0,0], color=:orange, markersize=4, label="Unstable equilibrium: $unstable_eq", legend=true)
scatter!(stable_eq, [0], color=:green, markersize=4, label="Stable equilibrium: $stable_eq", legend=true)


# Display the plot
savefig(p, "univariate_cubic_dyn_01jul24.png")
display(p)
```
